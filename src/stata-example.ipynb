Example:

```Macro.setGlobal("output_path", f"{OUTPUT_PATH}")``` calls the Stata Macro Stata API endpoint and sets a global variable (like ```global output_path path``` in Stata), setting it to a path set in the config.py file.

## Move data to Stata

The syntax at the beginning of the following cell is important. After the ```%%stata``` magic command (as it is called), we include some flags:

1. -d means that we want to send data to Stata from this notebook
2. df refers to the name of dataset we want to send
3. -force tells Stata to overwrite the current dataset in memory. Think of this like writing ```use df, clear```. Whether or not this is best practice is subjective, but since Stata is running "in the background", when we call it from Python, Stata never resets. This means that if you previously loaded data and didn't explicitly end the instance of Stata running in this notebook and therefore keeps any dataset previously loaded in memory.


````python
%%stata -d df -force
describe
````
